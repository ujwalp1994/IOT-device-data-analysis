```python
# Adding heartrate data to another column for prelimary visualizations 

test2 = pd.DataFrame()
test2 = pd.concat([test2, record_file1[['time [s/256]', 'heart_rate [bpm](/api/datatype/19/)']]], axis=0)
```

```python
# Cleaning out the Unwanted rows

test2.dropna()
```

```python
# Converting String data to datetime format

test2['time [s/256]'] =  pd.to_datetime(test2['time [s/256]'], format='%Y-%m-%d %H:%M:%S')
```

```python
test2.set_index('time [s/256]')
```

```python
# Getting avegrage breating rate hourly wise 

test3 = test2.reset_index().set_index('TimeSeries').resample('H').mean()
```

```python
# Importing libraries for visualization

import seaborn as sns 
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.plotly as pydata

import plotly.graph_objs as go
sns.set(style="whitegrid", color_codes=True)
plt.rc("font", size=10)
```

```python
# Renaming column names for better understanding

test2.rename(columns={'time [s/256]': 'TimeSeries', 'heart_rate [bpm](/api/datatype/19/)': 'heart_rate'}, inplace=True)
```

```python
test3
```

```python
# Taking variables for plotting data

heart_rate = test3['heart_rate']
series = test3.index
type(series)
```

```python
# Plotting the heart_rate based on the timeseries obtained

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot_date(x=series, y=heart_rate, fmt="r-")
plt.title("Heart Rate Ccomparison")
plt.ylabel("Heart Rate")
plt.xlabel("TimeSeries- Month-Day-Hour")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# The x and y data to plot
y = heart_rate
x = series

# Threshold above which the markers should be red
threshold1 = 100.0
threshold2 = 60

# Creating legend for Ideal heart_rate values
red_patch = mpatches.Patch(color='red', label='Above Ideal heart_rate range')
blue_patch = mpatches.Patch(color='blue', label='Normal heart rate')
green_patch = mpatches.Patch(color='green', label='Well-conditioned heart_rate')

# Create figure
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

# Labelling the graph
plt.title("Heart Rate Ccomparison")
plt.ylabel("Heart Rate")
plt.xlabel("TimeSeries- Month-Day-Hour")
plt.legend(handles=[red_patch, blue_patch, green_patch], loc=1)
plt.grid(True)
plt.tight_layout()

# Plot the line
plt.plot(x, y, color='blue')

# Add below threshold markers
below_threshold = y < threshold2
plt.scatter(x[below_threshold], y[below_threshold], color='green') 

# Add above threshold markers
above_threshold = y > threshold1
plt.scatter(x[above_threshold], y[above_threshold], color='red') 
```
